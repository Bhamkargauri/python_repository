2)	Write a program that keeps taking user input for password until ‘Enter’ key is pressed and then checks password with following validations:
    1)	At least 9 character
    2)	Contains at least 1 uppercase and 1 lowercase letter
    3)	Contains at least one number
    4)	Contains at least one special character and it should be one of the allowed special characters [$,_,%] 
    5)	If password is incorrect show related error else show a success message. 	


```python
allowed_special_characters = {'$', '_', '%'}

while True:
    password = input("Enter a password: ")
    
    if password == "":
        break

    if len(password) < 9:
        print("Password must be at least 9 characters long.")
        continue

    has_upper = False
    has_lower = False
    has_digit = False
    has_special = False

    for char in password:
        if char.isupper():
            has_upper = True
        if char.islower():
            has_lower = True
        if char.isdigit():
            has_digit = True
        if char in allowed_special_characters:
            has_special = True
 
    if not has_upper:
        print("Password must contain at least one uppercase letter.")
        continue
    if not has_lower:
        print("Password must contain at least one lowercase letter.")
        continue
    if not has_digit:
        print("Password must contain at least one number.")
        continue
    if not has_special:
        print("Password must contain at least one special character ($, _, %).")
        continue

    print("Password is valid!")
    break
```

```python
#Q.2> Write a query to get average salary of employees grouped by their department and show along with department name

SELECT d.Department_id, d.Department_name, AVG(e.Salary) AS Average_Salary
FROM employee_management.Department d
JOIN employee_management.Employees e ON d.Department_id = e.Department_id
GROUP BY d.Department_id, d.Department_name;
```

```python
#Q.3>Write a query to get the list of HOD’s whose salary is greater than average salary of HODs

SELECT e.Name, e.Salary
FROM employee_management.Employees e
JOIN employee_management.Department d ON e.Employee_id = d.HOD_id
WHERE e.Salary > (
    SELECT AVG(e.Salary)
    FROM employee_management.Employees e
JOIN employee_management.Department d ON e.Employee_id = d.HOD_id
);
```

```python
#Q.4> Write a query to get the names of youngest and oldest HOD with their age

SELECT Name, 
       YEAR(CURDATE()) - YEAR(DoB) AS Age
FROM employee_management.Employees
WHERE Employee_id IN (SELECT HOD_id FROM employee_management.Department)
ORDER BY Age ASC
LIMIT 1;  (it gives for top record)
```

```python
#Q.5> Write a query to get the list of departments where HOD’s earn at least twice the average salary of the other employees in that department.

SELECT d.Department_id, d.Department_name, e.Name AS HOD_Name, e.Salary AS HOD_Salary
FROM employee_management.Department d
JOIN employee_management.Employees e ON d.HOD_id = e.Employee_id
WHERE e.Salary >= 2 * (
    SELECT AVG(e2.Salary)
    FROM employee_management.Employees e2
    WHERE e2.Department_id = d.Department_id
      AND e2.Employee_id != d.HOD_id  );
```
